```python
##Importing important libraries
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
com=pd.read_csv('Company_Data (1).csv')
com.head()
```

```python
com.describe()
```

```python
com["sales"]=(com["Sales"]>7.49)*1
```

```python
com
```

```python
label_encoder=preprocessing.LabelEncoder()
com["ShelveLoc"]=label_encoder.fit_transform(com["ShelveLoc"])
com["Urban"]=label_encoder.fit_transform(com["Urban"])
com["US"]=label_encoder.fit_transform(com["US"])
```

```python
com
```

```python
x=com.iloc[:,1:9] ##11
```

```python
y=com.iloc[:,-1].values
```

```python
from sklearn.model_selection import KFold
```

```python
kfold=KFold(n_splits=10,random_state=42, shuffle= True)
```

```python
kfold
```

```python
model = RandomForestClassifier(n_estimators=110)
```

```python
result=cross_val_score(model,x,y,cv=kfold)
print(result.mean())
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
model.fit(x_train,y_train)
```

```python
from sklearn import metrics
```

```python
Y_pred=model.predict(x_train)
```

```python
print("accuracy:",metrics.accuracy_score(y_train,Y_pred))
```

```python
model.score(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
model.score(x_test, y_test)
```

```python
model.predict([[110,100,4,466,97,2,55,14,]])
```

```python
feature_imp = pd.Series(model.feature_importances_,index=x.columns).sort_values(ascending=False)
feature_imp
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Creating a bar plot
sns.barplot(x=feature_imp, y=feature_imp.index)
# Add labels to your graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
plt.show()
```

```python
train_accuracy = []
test_accuracy = []
for max_features in range(1,8):
    model=RandomForestClassifier(n_estimators=110, max_features=max_features)
    model.fit(x_train, y_train)
    train_accuracy.append(model.score(x_train, y_train))
    test_accuracy.append(model.score(x_test, y_test))
```

```python
data = pd.DataFrame({'max_features':range(1,8),'train_acc':train_accuracy, 'test_acc':test_accuracy})
data
```

```python
# visulizing accuracy graph to look over max_depth
labels =['train_acc','test_acc']
fig = plt.figure(figsize=(15,6))
ax = fig.add_subplot()
plt.plot(data['max_features'],data['train_acc'], marker = 'o')
plt.plot(data['max_features'], data['test_acc'], marker = 'o')
ax.set_xticks(np.arange(1,10))
plt.legend(labels)
plt.xlabel('max_features of the tree')
plt.ylabel('performance')
plt.show()
```
